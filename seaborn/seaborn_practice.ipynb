# 探索变量同关系（基于Sesborn）
- 数据变量之间的关联性，主要针对定量数据而言；
- 数据的分布问题也是主要针对定量数据；
- 分组问题及组间问题里，就是在数据分析中，一定会包括定序和定类数据。

```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()

# Setting a random seeds
np.random.seed(sum(map(ord, 'regression')))

tips = sns.load_dataset('tips') #Restaurant Tips Case
tips.head()
```

## 绘制线性回归模型
- 散点图+线性回归+95%置信区间
- 小费收入于餐厅消费成正比

```python
sns.lmplot(x='total_bill', y='tip', data=tips) # Confidence interval
```

如图所示，帐单总量与小费之间存在线性相关关系

```python
sns.lmplot(x='size', y='tip', data=tips)
```

离散数据，并不太适合

```python
# 方法一： 增加抖动
sns.lmplot(x='size', y='tip', data=tips, x_jitter= 0.2)
```

```python
#  用均值和置信区间代替离散点
sns.lmplot(x='size', y='tip', data=tips, x_estimator= np.mean)
```
